```python
x = df[['age','sex','bmi','children','smoker']]
y = df['charges']
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=42)
```

```python
def outlier_count(col, data = x_train):
    
    # QD
    q75, q25 = np.percentile(data[col], [75, 25])
    # inter quatile
    iqr = q75 - q25
    # min_val and max_val
    min_val = q25 - (iqr*1.5)
    max_val = q75 + (iqr*1.5)
    
    # outlier count
    outlier_count = len(np.where((data[col] > max_val) | (data[col] < min_val))[0])
    # outlier in percent
    outlier_percent = round(outlier_count/len(data[col])*100, 2)
    
    if(outlier_count > 0):
        print("\n"+15*'-' + col + 15*'-'+"\n")
        print('Number of outliers: {}'.format(outlier_count))
        print('Percent of data that is outlier: {}%'.format(outlier_percent))
```

```python
for col in x_train.columns:
    outlier_count(col, x_train)
```

```python
print(f"Shape of x_train: {x_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of x_test: {x_test.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lr = LinearRegression()
lr.fit(x_train, y_train)
predictData = lr.predict(x_test)

print(f"Mean Square Error is {mean_squared_error(y_test, predictData)}")
print(f"R square is {r2_score(y_test, predictData)}")
```

```python
# cross validation + grid search
from sklearn.linear_model import LinearRegression  
# we will be using linaer regression because we have less error cause R square is too close to 1.
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

algorithms = [
    LinearRegression(),
    SVR(),
    KNeighborsRegressor(),
    DecisionTreeRegressor(random_state= 0),
    RandomForestRegressor(n_estimators = 100, random_state = 0)
]

algorithmNames = [
    "Linear Regression",
    "Support Vector Regression",
    "K-Neighbors Regression",
    "Decision Tree Regression",
    "Random Forest Regression"
]
```

```python
from sklearn.model_selection import KFold, cross_val_score

train_mse = []
test_mse = []

kfold = KFold(n_splits=5, shuffle=True)

for i, model in enumerate(algorithms):
    scores = cross_val_score(model, x_train, y_train, cv=kfold, scoring='neg_mean_squared_error')
    print(f"{algorithmNames[i]} - Score: {scores}; Mean: {scores.mean()}")
```

```python
# Using random forest because Mean score is Lowest in RandomForest
# Grid Search 

from sklearn.model_selection import GridSearchCV

param_grid = {'bootstrap': [True], 'max_depth': [5, 10, None],
              'n_estimators': [5, 6, 7, 8, 9, 10, 11, 12, 13, 15]}

rf = RandomForestRegressor(random_state = 1)

grid = GridSearchCV(estimator = rf, 
                    param_grid = param_grid, 
                    cv = kfold, 
                    n_jobs = -1, 
                    return_train_score=True, 
                    refit=True,
                    scoring='neg_mean_squared_error')

grid.fit(x_train, y_train); 
```

```python
grid.best_params_
```

```python
best_mse = grid.best_score_
```

```python
best_mse
```

```python
# 7. Testing 

predictData = grid.predict(x_test)
mse = mean_squared_error(y_test, predictData)
r2 = r2_score(y_test, predictData)
print(f"MSE: {mse}")
print(f"R2: {r2}")
```
